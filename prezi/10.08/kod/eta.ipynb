```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def eta(t):
    return -np.log(np.tan((np.radians(t/2))))
```

```python
t = np.arange(0.0, 45.)
eta = np.vectorize(eta)
plt.figure()
# Plotting the correlations of various variables with mass along with the errors
fig, ax = plt.subplots(figsize=(15,10))
ax.set_xlabel(r'$\theta$ ($\degree$)', fontsize=15, loc='right')
ax.set_ylabel(r'$\eta =-\ln(\tan(\theta)/(2))$', fontsize=15, loc='top')
plt.plot(t, eta(t))
plt.title('Pseudorapidity', fontsize=20)
#plt.tight_layout()
fig.savefig('eta.png')#, bbox_inches='tight')
```
